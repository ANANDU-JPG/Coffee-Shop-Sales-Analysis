```python
co.dtypes
```

```python
co['datetime'] = pd.to_datetime(co['datetime'])

# Extract time part and create a new 'time' column
co['time'] = co['datetime'].dt.time

print(co)
```

```python
co.head()
```

```python
co.shape
```

```python
co.isnull().sum()
```

```python
co.dropna(subset=["card"],axis=0,inplace=True)
co
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
co["coffee_name_e"]=le.fit_transform(co["coffee_name"])
co['cash_type_e']=le.fit_transform(co['cash_type'])
co['card_e']=le.fit_transform(co['card'])
co['datetime_e']=le.fit_transform(co['datetime'])
co
```

```python
co.head()
```

```python
sns.set(style="whitegrid")

# --- Plot 1: Distribution of Coffee Types ---
plt.figure(figsize=(10,4))
sns.countplot(data=co, x='coffee_name', palette='pastel')
plt.title("Coffee Type Distribution")
plt.xlabel("Coffee Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=co, x='coffee_name', y='money', estimator=sum, ci=None, palette='muted')
plt.title("Total Revenue by Coffee Type")
plt.xlabel("Coffee Type")
plt.ylabel("Total Revenue")
plt.tight_layout()
plt.show()
```

```python
co['hour'] = pd.to_datetime(co['datetime']).dt.hour
plt.figure(figsize=(6,4))
sns.countplot(data=co, x='hour', palette='coolwarm')
plt.title("Transactions by Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Transactions")
plt.tight_layout()
plt.show()
```

```python
customer_counts = co['card'].value_counts()
returning_customers = (customer_counts > 1).sum()
print(f"\nReturning Customers: {returning_customers} / {co['card'].nunique()}")
```

```python
co['minute'] = pd.to_datetime(co['datetime']).dt.minute
plt.figure(figsize=(6, 3))
sns.heatmap(co[['money', 'hour', 'minute']].corr(), annot=True, cmap='Blues')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
co['time_only'] = co['datetime'].dt.time

plt.figure(figsize=(12,6))
sns.stripplot(x='coffee_name', y=co['datetime'].dt.hour + co['datetime'].dt.minute/60, data=co, jitter=True, palette='Set2')
plt.title('Transaction Time (Hour) by Coffee Type')
plt.xlabel('Coffee Type')
plt.ylabel('Time of Day (Hours)')
plt.show()
```

```python
avg_price_by_hour = co.groupby('hour')['money'].mean().reset_index()

plt.figure(figsize=(8,4))
sns.lineplot(x='hour', y='money', data=avg_price_by_hour, marker='o', color='blue')
plt.title('Average Price of Orders by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Average Price')
plt.xticks(avg_price_by_hour['hour'])
plt.show()
```

```python
pivot = co.pivot_table(index='hour', columns='coffee_name', values='money', aggfunc='count', fill_value=0)

plt.figure(figsize=(8,6))
sns.heatmap(pivot, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Number of Transactions by Hour and Coffee Type')
plt.xlabel('Coffee Type')
plt.ylabel('Hour of Day')
plt.show()
```

```python
hourly_items = co.groupby(["hour", "item"])["customer_id"].count().reset_index()
hourly_items = hourly_items.rename(columns={"customer_id": "purchase_count"})

plt.figure(figsize=(14,6))
sns.lineplot(x="hour", y="purchase_count", hue="item", data=hourly_items)
plt.title("Most Purchased Coffee by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Purchases")
plt.legend(title="Coffee Type", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
# Calculate the correlation matrix for numerical columns
correlation_matrix = co[['money', 'hour', 'minute', 'coffee_name_e', 'card_e', 'cash_type_e', 'datetime_e']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```
